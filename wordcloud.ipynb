```python
a = '注意力、资源、软件、app、眼睛、记忆、网络、作业、疲劳、笔记、电子产品、自觉性、课堂、任务、笔记、走神、遗忘、错过、混乱、干扰、交流、效率、设备、课本、自制力、枯燥、深度、碎片、方法、环境、疑问'
```

```python
a
```

```python
a = a.replace('、',' ')
```

```python
lis = [a.split(' ')[i] for i in range(31)]
```

```python
docname = "doc.txt"
with open(docname) as f:
    doc = f.read()
```

```python
doc.count(lis[0])
```

```python
tim = []
for i in lis:
    tim.append([i, doc.count(i)])
```

```python
final = []
for i in range(len(lis)):
    final = final + [tim[i][0]] * tim[i][1]
```

```python
finall = ''
for x in final:
    finall = finall + ' ' + x
```

```python
from wordcloud import WordCloud
wordcloud = WordCloud(font_path="SimHei",collocations=False,width=900,height=400,background_color='orange',max_words=100,scale=1.5).generate(finall)
%pylab inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=  (15,8)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
#plt.savefig('wordcloud.png')
```
